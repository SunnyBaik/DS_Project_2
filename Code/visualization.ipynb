```python
path = '/Users/jonghyunchoe/Documents/Others/GitHub/DS_Project_2/CSV/'
predicted_train = pd.read_csv(path + 'predicted_train.csv')
model_predicted_train = pd.read_csv(path + 'model_predicted_train.csv')
predicted_test = pd.read_csv(path + 'predicted_test.csv')
```

```python
# plot의 사이즈를 지정하는 코드로  원하는 사이즈로 설정하면 됩니다.
plt.rcParams["figure.figsize"] = (18,5) 

def plot_time(df, time=(0, 100), columns='all'):
    df_time = df[(df['time']>=time[0]) & (df['time']<time[1])]
    column_list = df.columns if columns=='all' else columns

    for col in column_list:
        if col !='time':
            plt.plot('time', col, data=df_time)

    plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5))
    plt.show()
```

```python
# plot_time(train, time=(100, 500))
predicted_train['time'] = predicted_train.index/100
plot_time(predicted_train, time=(10000, 11000), columns=['s1', 'label', 'predicted_label'])
```

```python
model_predicted_train['time'] = model_predicted_train.index/100
plot_time(model_predicted_train, time=(0, 10000), columns=['s1', 'label', 'predicted_label'])
```
